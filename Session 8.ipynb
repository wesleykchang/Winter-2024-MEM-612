## Boiling

A classic example: water boiling involves a phase change from saturated liquid water to water vapor

1. Natural convection: below the temperature at which gas bubbles start to form, natural convection is the primary mode of heat transfer as water is heated

2. Nucleate boiling: here, nucleation sites cause isolated bubbles to form and rise. As bubbles become more concentrated, they start to merge and form jets and slugs of vapor

3. Eventually, a peak heat flux is reached, beyond which the heat flux diminishes. This is because the vapor escape process becomes less effective as there is too much vapor that prevents further gas from escaping. 

4. Transitional boiling regime and film boiling. The heat is now conducted through a vapor film instead of the liquid film. The heat transfer coefficients are subsequently less than before. 





### Nucleate boiling 

1. Isolated nucleation sites lead to bubbling. As q and $\Delta T$ increase, more sites are activated.

2. Slugs and columns. Bubbles start to merge into jets of vapor.

### Peak boiling heat flux 

Peak heat flux region has the highest rate of heat transfer, but can be dangerous in practical sense because if exceeded can lead to large temperature rise

The heat flux diminishes beyond the peak, because a large  vapor blanket forms which decreases heat transfer efficacy.

## Surface tension



Vapor pockets can be idealized as spherical bubbles of pure saturated steam.

If the bubble is in equilibrium with a surrounding superheated liquid, then the temperature will be uniformly the superheated temperature throughout.

The saturated vapor pressure inside should be slightly higher than ambient since the surface tension adds to the pressure.

The surface tension of water in contact with vapor is: $\sigma _{water}\ =\ 235.8 (1\ -\ T_{sat}/T_c)^{1.256}[1\ -\ 0.625(1\ -\ T_{sat}/T_c)]$ 

where $T_c$ is the critical point temperature and the units for surface tension are in mN/m. 

if the bubble's radius is less than an equilibrium value, surface tension will dominate and the bubble will collapse (at this higher pressure, the vapor will condense into a liquid)

if the bubble's radius is larger than equilibrium, the bubble will grow as the liquid outside evaporates 

## Peak pool boiling heat flux

Boiling water rises upward. On the flip side, a heavy fluid (honey in a jar) sinks, in a similar manner.

The movement is wave-like, either collapsing upwards or downwards, and is termed Taylor instability. the wavelength is related to surface tension, inertia and gravity.
