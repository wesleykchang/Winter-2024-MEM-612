b) The general equation is: $u\ =\ r^2/4\mu (dP/dx)\ +\ c_1 ln(r)\ +\ c_2$

c) plugging in boundary conditions and solving for the two constants

$c_1\ ln(r_o/r_i)\ =\ -1/4\mu dP/dx(r_o^2\ -\ r_i^2)$

$\tau _o\ =\ -\mu (du/dr)_{r=r_o}$

if the inner tube shrinks, the difference in the shear stresses should increase. plotting (see below), we see that a decreasing inner radius increases the shear stress of both walls (which is directed in the reverse direction of flow), with the shear stresss of the inner wall increasing more.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

r_i = np.linspace(1,10,10)

tau_i = r_i/2*1 - 0.25/r_i * (20**2 - r_i**2)/np.log(20/r_i)

tau_o = -20/2 * 1 + 1/80 * (20**2 - r_i**2)/np.log(20/r_i)

plt.plot(r_i, tau_i, c = 'b')

plt.plot(r_i, tau_o, c = 'g')

plt.show()
```
