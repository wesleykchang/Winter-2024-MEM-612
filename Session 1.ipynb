## Why is heat transfer important?

Regardless of where you are in the known universe, the total energy of an arbitrary system is conserved (First Law of Thermodynamics), and therefore, for any given system, there are only two ways energy can flow through the boundaries: heat transfer through the boundaries, or work done on or by the system. $\Delta E\ =\ Q\ -\ W$

We live in a world of fundamentally limited resources. Unless we're importing things in from outerspace, we will always inhabit a planet that has finite resources. Over the years, we have gotten more and more efficient with how we use these resources. Whether it's building internal combustion engines, or power plants, or batteries and other renewable forms of energy storage, our societies today rely on efficient heat transfer engineering. Ultimately, technologies that consume less fuel or energy to produce a given amount of work will allow us to use our limited resources more efficiently, and prolong our existence on Earth. 

In particular, many forms of heat transfer both natural or man-made rely on *convection* or transfer of heat due to bulk fluid motion. Air and ocean currents, refrigeration, heat exchangers, building HVAC systems, rocketry, battery thermal management systems, etc.


## What is convection?

MEM 611 was about *conductive* heat transfer, or heat transfer that occurs across a medium (without any moving fluids). Convective heat transfer, on the other hand, occurs via *bulk fluid motion*. In other words, heat moves *together* with moving fluid. 

Specifically, the fluid immediately next to a body forms a thin slowed-down region called a boundary layer, and heat from the surface is conducted into this layer, getting swept into the bulk fluid. 

## Common heat transfer boundary conditions

The temperature distribution / solution will depend on the boundary conditions, and for time-dependent (transient) problems, the initial condition. Here are some common boundary conditions:

1) constant surface temp $T(0,t)\ =\ T_s$
2) constant surface heat flux $-k\frac{dT}{dx}|_{x=0}\ =\ q_s"$ (=0 if insulated surface)
3) convective surface $-k\frac{dT}{dx}|_{x=0}\ =\ h[T_\infty\ -\ T(0,t)]$


## Boundary layers

When a fluid moves past a surface, it "sticks", leading to variable velocities near the surface. This effect disappears further away from the surface. Similarly, if the surface temperature and the fluid stream temperature are different, then a temperature gradient will form. This layer, where the temperature gradient occurs, is called the *boundary layer thickness*. Usually, the edge of the boundary layer is defined as the location where the temperature is about 99% that of the bulk fluid temperature. The boundary layer has a defined thickness, usually where velocity is 99% of the bulk fluid velocity. 





At any distance x away from the leading edge of the moving fluid, the surface heat flux is defined by Fourier's law (for the fluid):

$$q_s"\ =\ -k_f\frac{dT}{dy}|_{y=0} $$

This holds true because at the surface, heat transfer occurs solely by conduction. Remember, convection is heat transfer due to the bulk moving fluid. 

Newton's law of cooling then relates this surface heat flux to the difference in temperatures between the surface and the environment, with a constant of proportionality h defined as the heat transfer coefficient (W/m2-K). 

What is h dependent on? 
- properties of the fluid, such as density, thermal conductivity, specific heat, viscosity
- surface roughness 

$$-k_f \frac{dT}{dy}|_{y=0}\ =\ h(T_w\ -\ T_\infty) $$

Fourier's law of conduction (lefthand side) describes heat removal at wall because velocity at the surface is 0. 

The thermal boundary layer non-dimensional variable is the Nusselt number (Nu), or $Nu_L\ =\ \frac{hL}{k_f}$ since $\frac{\delta (T_w-T/T_w-T_\infty)}{y/L}\ =\ hL/k_f$. The Nusselt number describes the ratio of convective to conductive heat transfer. Therefore, a larger Nu corresponds to more convective heat transfer (100 to 1000), whereas a smaller Nu corresponds to laminar (1 to 10) flow or pure conduction (1). 

Lastly, this boundary layer also exists for concentration gradients (mass transport). In that case, we refer to Fick's law of diffusion.

## Lumped capacity solution 

Now consider a body with almost uniform internal temperature. When internal temp gradients are small, we can consider T = T(t)

We started the First Law of Thermo: $Q\ =\ dU/dt$

$-\overline{h}A(T-T_\infty)\ =\ \frac{d}{dt}(\rho cV(T\ -\ T_ref))$

$\frac{d(T-T_\infty)}{dt}\ =\ -\frac{\overline{h}A}{\rho cV}(T\ -\ T_\infty)$

$ln(T-T_{\infty})\ =\ -\frac{t}{\rho cV/\overline{h}A}\ +\ c$

if initial temp is $T(t=0) = T_i$, then $c\ =\ ln(T_i\ -\ T_\infty)$

$\frac{T\ -\ T_\infty}{T_i\ -\ T_\infty}\ =\ e^{-t/ \tau}$ where $\tau$ is the time constant $\rho cV/\overline{h}A$

a heavier mass will take longer to cool (larger time constant), i.e. more resistive to heat convection 

in this, thermal conductivity is neglected, bcs we assume internal conduction is not important. in other words the Biot number is very low (Bi = $\frac{\overline{h}}{k_b}$) which implies $T_{bulk}\ =\ T_{surface}$

if the Biot number were larger, then the reverse is true. Convective heat transfer can be neglected and we focus on the heat conduction equation subject to simple boundary conditions. in other words, the Biot number tells us how to approach the problem - either from a convective or a conductive point of view. 


## Laminar and turbulent flow

Laminar fluid flow is highly ordered. A transition period can exist prior to turbulent flow. The onset of turbulent flow is dependent on the Reynolds number, a dimensionless number that relates inertial with viscous forces. 

$$Re_x\ =\ \frac{\rho u_\infty x}{\mu}\ =\ \frac{u_\infty x}{v}$$

x is a characteristic distance from the leading edge (along the flow path). $u_\infty$ is the fluid viscosity, v is the kinematic viscosity $\mu /\rho$. Over a flat plate, Re can vary between 1E5 and 3E6. Laminar flow can persist up to ~10,000. 

$\delta / x\ =\ 4.92/\sqrt{Re_x}$ was the result of Ludwig Prandtl's analysis. In other words, large Reynolds numbers lead to thin boundary layer thicknesses. However, near the leading edge or when fluid moves slowly, Re may be < 1000 and this relationship may break down.

Generally, high velocities or low viscosities lead to thinner boundary layers, which offer less resistance to heat flow and increases heat convection. 

Consider fluid flow in a pipe. The max Reynolds number for laminar flow in a pipe is roughly 2100. $Re_{critical}\ =\ \frac{u_\infty x}{v}$

## Laminar incompressible boundary layer on a flat surface

The conservation of mass and momentum in the boundary layer can be solved to predict the boundary layer flow field. Within laminar flow, there are streamlines or defined pathways of bulk movement, no fluid flows across these streamlines.

The mass flow rate between the streamlines is $d\dot{m}\ =\ \rho \overrightarrow{u}*\overrightarrow{n} dA\ =\ \rho d\phi$ where $phi$ is the stream function ($\phi(x,y)\ =\ constant$). 

1) First, consider an area on a streamline $dA$, with a normal vector $\overrightarrow{n_s}$. The mass flow is $d\dot{m}\ =\ \rho \overrightarrow{u}*\overrightarrow{n_s}dA\ =\ 0$

2) $d\dot{m}\ =\ -\rho v dx\ +\ \rho u dy\ =\ 0$

since $\overrightarrow{n_s}dA\ =\ -dy \overrightarrow{i}\ +\ dx \overrightarrow{j}$

$d\phi\ = \frac{d\phi}{dx}|y dx\ +\ \frac{d\phi}{dy}|x dy\ =\ 0$

3) by substituting stream functions into the velocities:

$v \ =\ -\frac{d\phi}{dx}|y$, $u\ =\ \frac{d\phi}{dy}|x$

since $\frac{d^2\phi}{dydx}\ =\ \frac{d^2\phi}{dxdy}$

$\frac{du}{dx}\ +\ \frac{dv}{dy}\ =\ 0$

This is the two dimensional continuity equation for incompressible flow. In other words, the mass that enters a volume of fluid must also leave it. 




##  Example: fluid with uniform velocity in x-direction

Fluid moves at a constant velocity $u_\infty$, v = 0. 

$\frac{d\phi}{dy}|_x\ =\ u_\infty$, $v\ =\ -\frac{d\phi}{dx}|_y\ =\ 0$

Integrating:

$\phi\ =\ u_\infty y\ +\ f(x)$, $\phi\ =\ 0\ +\ f(y)$

therefore, f(x) = constant and $f(y)\ =\ u_\infty y\ +\ c_1$

$\phi\ =\ u_\infty y\ +\ c_1$

## What about momentum? 

Particles in the fluid feel shear stresses which create momentum. 




$u \frac{du}{dx}\ +\ v\frac{du}{dy}\ =\ -\frac{1}{p}\frac{dp}{dx}\ +\ v\frac{d^2u}{dy^2}$

There are several boundary layer approximations:

- $|du/dx| << |du/dy|$, in other words: shear stresses in the x-direction dominate 
- v << u
- y pressure gradient is negligible

## Energy Equation

Now, we extend the heat conduction equation to allow for fluid motion: $q\ =\ \frac{q}{T_w\ -\ T_\infty}\ =\ -\frac{k}{T_w\ -\ T_\infty}\frac{dT}{dy}|_{y=0}$



Consider an arbitrarily shaped volume under a temperature field and containing a moving fluid with velocity $\overrightarrow{u}(x,y,z)$

$\frac{d}{dt}\int _R \rho \hat{u} dR\ =\ -\int _S (\rho \hat{h})\overrightarrow{u}*\overrightarrow{n}dS\ -\ \int _S (-k \nabla{T})*\overrightarrow{n}dS\ +\ \int _R \dot{q} dR$

rate of internal energy increase in R is equal to the rate of internal energy and flow out of R minus net heat conduction out of R plus heat generation in R. 

R is the shape, dS is an element on the surface, dR is a volume element inside the shape. $\hat{u}$ is the specific internal energy, $\hat{h}$ is the specific enthalpy. the second term is a volume flow rate $\overrightarrow{u}*\overrightarrow{n}dS$ through surface element

recall Gauss' theorem to make surface integrals into volume integrals

$\int _R (\frac{d(\rho \hat{u})}{dt}\ +\ \nabla{\rho \overrightarrow{u}\hat{h}}\ -\ \nabla k\nabla T\ -\ \dot{q})dR\ =\ 0$

usually, k depends weakly on T (remember plots from MEM611). and a few other approximations are made..(derivatives of density are negligible if density changes are small)

$\rho (\frac{d\hat{h}}{dt}\ +\ \overrightarrow{u}*\nabla \hat{h})\ +\ \hat{h}(\frac{d\rho}{dt}\ +
 \nabla *(\rho \overrightarrow{u}))\ =\ k\nabla ^2 T\ +\ \dot{q}$

 substitute $d\hat{h}\ =\ c_p dT$

 the energy equation is: $\rho c_p (\frac{dT}{dt}\ +\ \overrightarrow{u}*\nabla T)\ =\ k\nabla ^2 T\ +\ \dot{q}$

energy storage + enthalpy convection term = heat conduction + heat generation 

the term in parenthesis on the left hand side is the material derivative D/Dt, which describes the rate of change of temperature of a fluid particle under movement in a flow field:

$dT/dt\ +\ \overrightarrow{u}*\nabla T\ =\ \frac{dT}{dt}\ +\ u\frac{dT}{dx}\ +\ v\frac{dT}{dy}\ +\ w\frac{dT}{dz}\ =\ \frac{DT}{Dt}$

in a steady state flow field without heat sources: $u\frac{dT}{dx}\ +\ v\frac{dT}{dy}\ =\ \alpha (\frac{d^2 T}{dy^2}\ +\ \frac{d^2T}{dx^2})$



## Prandtl number and a quick take on microkinetic theory of gases

Prandtl number = $\frac{v}{\alpha}\ =\ \frac{\mu c_p}{k}$ 

what is the physical meaning of this number? 

1) consider a bunch of gas molecules colliding with each other. the mass flux is proportional to $\rho \overline{C}$ where $\overline{C}$ is the mean speed

$q_y\ =\ c_2(\rho c_v \overline{C})(l\frac{dT}{dy})$

$k\ =\ c_2(\rho c_v \overline{C}l)$ so $\alpha\ =\ c_2 \frac{\overline{C}l}{\gamma}$

Pr is a constant on the order of one.

- for simple monatomic gases, around 2/3

- for diatomic molecules, 5/7

- more complex molecules, Pr gets closer to 1 

How does gas composition influence sensitivity to temperature? 
Pr is least sensitive to temperature in simple gases bcs structure is least responsive to temperature change

in liquids, Pr could be very far from 1, liquid metals very low, hydrocarbons 1E5. 

for Pr = 1, the dimensionless velocity and temperature profiles are identical on a flat surface 

2) Pr is the ratio of kinematic viscosity (v) to thermal diffusivity (alpha), and measures the relative effectiveness of momentum and energy transport by diffusion in the velocity and thermal boundary layers, respectively. in other words, gases have low Pr and viscous fluids have high Pr values, as they experience much larger velocity boundary layers than thermal boundary layers

$\frac{\delta}{\delta _t}\ \sim\ Pr^n$
